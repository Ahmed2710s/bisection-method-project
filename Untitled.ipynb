```python
def bisection_method(f, a, b, tol=1e-6, max_iter=100):
    """
    Bisection method to find root of f(x) = 0
    """
    if f(a) * f(b) >= 0:
        raise ValueError("f(a) and f(b) must have opposite signs")

    print(f"{'Iteration':<10} {'a':<10} {'b':<10} {'c':<10} {'f(c)':<12} {'Error':<10}")
    print("-" * 70)

    for i in range(max_iter):
        c = (a + b) / 2
        fc = f(c)
        error = abs(b - a)

        print(f"{i+1:<10} {a:<10.6f} {b:<10.6f} {c:<10.6f} {fc:<12.6f} {error:<10.6f}")

        if abs(fc) < tol or error < tol:
            print(f"\nConverged after {i+1} iterations!")
            return c, i + 1

        if f(a) * fc < 0:
            b = c
        else:
            a = c

    print(f"\nMaximum iterations reached")
    return (a + b) / 2, max_iter

# Example usage
if __name__ == "__main__":
    def f(x):
        return x**3 - 2*x - 5

    print("BISECTION METHOD: f(x) = x³ - 2x - 5")
    root, iterations = bisection_method(f, 2, 3)
    print(f"\nFinal root: {root:.8f}")
```
